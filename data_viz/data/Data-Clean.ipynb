```python
a=combined.groupby('What US state or territory do you work in?').mean()
a=pd.Series(a['What is your age?'].values,index=a.index).to_dict()
for key, value in a.items():
    a[key]=round(value,2)
import json

with open('/Users/pranavmahajan/documents/Data-Vis Project/us-states.js') as dataFile:
    data = dataFile.read()
    obj = data[data.find('{') : data.rfind('}')+1]
    jsonObj = json.loads(obj)
```

```python
states = {}
count=0
for item in jsonObj['features']:
    count+=1
    if count<8:
        it=str(item['id'])
        it=it[1:]
        it=int(it)
    else:
        it=item['id']
    states[item['properties']['name']]=it
current=combined.groupby(['What US state or territory do you work in?','Do you currently have a mental health disorder?'])['What is your gender?'].count()
state='Alabama'
d={}
tot=0
yes=0
count=0
for index,value in current.items():
    print(index[0])
    
    if state != index[0]:
        d[state]= (yes/tot)
        yes=0
        tot=0
        state=index[0]
    if index[1]=='Yes':
        yes=value
    tot+=value
d['Wisconsin']=0.64
```

```python
prior=combined.groupby(['What US state or territory do you work in?','Have you had a mental health disorder in the past?'])['What is your gender?'].count()
```

```python
state='Alabama'
dd={}
tot=0
yes=0
count=0
for index,value in prior.items():    
    if state != index[0]:
        dd[state]= (yes/tot)
        yes=0
        tot=0
        state=index[0]
    if index[1]=='Yes':
        yes=value
    tot+=value
dd['Wisconsin']=0.71
for item in a:
    #print(item)
    l=[]
    l.append(states[item])
    l.append(a[item])
    l.append(d[item])
    l.append(dd[item])
    a[item]=l
```

```python
data_state=pd.DataFrame.from_dict(a,orient='index').rename(columns={0:'id',1:'Average Age',2:'Currently Diagnosed',3:'Previously Diagnosed'})
data_state.reset_index(inplace=True)
data_state.rename(columns={'index':'State'},inplace=True)
data_state.to_csv('/Users/pranavmahajan/documents/Data-Vis Project/age.tsv',sep = '\t')
```
